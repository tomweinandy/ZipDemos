```python
df_household_composition = pd.read_csv('D02.csv', header=1)
df_household_composition = df_household_composition.transpose()

# Set the first row as the column headers (then drop
df_household_composition.columns = df_household_composition.iloc[1]
df_household_composition = df_household_composition[1:]

q_kids = 'Percent!!HOUSEHOLDS BY TYPE!!Total households!!Households with one or more people under 18 years' #iloc[336]
q_spanish = 'Percent!!LANGUAGE SPOKEN AT HOME!!Population 5 years and over!!Spanish' #.iloc[540]
q_adult = 'Percent!!VETERAN STATUS!!Civilian population 18 years and over' #.iloc[446]

# df_kids = df_household_composition[df_household_composition['GEO_ID']==q_kids]
df_kids = df_household_composition[df_household_composition.index==q_kids]
df_kids['Variable'] = 'Households with kids'
#
# df_spanish = df_household_composition[df_household_composition['GEO_ID']==q_spanish]
df_spanish = df_household_composition[df_household_composition.index==q_spanish]
df_spanish['Variable'] = 'Spanish speaking household'
#
# df_adults = df_household_composition[df_household_composition['GEO_ID']==q_adult]
df_adults = df_household_composition[df_household_composition.index==q_adult]
df_adults['Variable'] = 'Adult population'

df2 = pd.concat([df_kids, df_spanish])
df2 = pd.concat([df2, df_adults])
# df2 = df2.reset_index()
#
df2
```
